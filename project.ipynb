```python
df=df[df['year'].str.isnumeric()]
```

```python
df['year'].unique()
```

```python
df['name'] = df['name'].str.split().str[:3].str.join(' ')
```

```python
df.describe()

df.dropna(inplace=True)
df.isnull().sum()
df.shape
df.describe()
```

```python

df = pd.get_dummies(df, columns=['company'], drop_first=True)

bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df['Price'].value_counts()
```

```python
df['year'].isna().sum()
```

```python
df.head()
```

```python
df['kms_driven'].head()
df['kms_driven'] = df['kms_driven'].str.replace(' kms', '', regex=False)
```

```python
df.info()
```

```python

df=df[df['year'].str.isnumeric()]
df['year']= df['year'].astype(int)
```

```python
df['Price'] = (
    df['Price']
    .str.replace(',', '', regex=False)
)

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
df['kms_driven'] = (
    df['kms_driven']
    .str.replace(' kms', '', regex=False)
    .str.replace(',', '', regex=False)
)

df['kms_driven'] = pd.to_numeric(df['kms_driven'], errors='coerce')
```

```python


df['Price']=df['Price'].fillna(df['Price'].mean())

df['kms_driven']=df['kms_driven'].fillna(df['kms_driven'].mean())


df['fuel_type'] = df['fuel_type'].fillna(df['fuel_type'].mode()[0])
df.isnull().sum()


```

```python

df = pd.get_dummies(df, columns=['fuel_type'], drop_first=True)
```

```python
X=df.drop(columns='Price')
Y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, Y,
test_size=0.2,random_state=42     # for reproducibility
)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
model = LinearRegression()
df.head()
df.drop(columns=['name'],inplace=True)
```

```python


df['fuel_type_LPG'] = df['fuel_type_LPG'].astype(int)

df['fuel_type_Petrol'] = df['fuel_type_Petrol'].astype(int)
df.head()
```

```python
model.fit(X_train,y_train)
y_pr=model.predict(X_test)
```

```python
df.head()
y_test.shape
X = X.reset_index(drop=True)
```
